```python
name = input("Enter student name: ")

m1 = float(input("Enter marks of Subject 1: "))
m2 = float(input("Enter marks of Subject 2: "))
m3 = float(input("Enter marks of Subject 3: "))
m4 = float(input("Enter marks of Subject 4: "))
m5 = float(input("Enter marks of Subject 5: "))

total = m1 + m2 + m3 + m4 + m5
average = total / 5

if average >= 75:
    grade = "A"
elif average >= 60:
    grade = "B"
elif average >= 50:
    grade = "C"
elif average >= 40:
    grade = "D"
else:
    grade = "Fail"

print("Student Name:", name)
print("Total Marks:", total)
print("Average Marks:", average)
print("Grade:", grade)
```

```python
print("----- Electronic Tools Menu -----")
print("1. Resistor  (₹2 each)")
print("2. Capacitor (₹5 each)")
print("3. Diode     (₹3 each)")
print("4. Transistor(₹10 each)")

choice = int(input("Enter your choice (1-4): "))
quantity = int(input("Enter quantity: "))

if choice == 1:
    total_cost = quantity * 2
    item = "Resistor"
elif choice == 2:
    total_cost = quantity * 5
    item = "Capacitor"
elif choice == 3:
    total_cost = quantity * 3
    item = "Diode"
elif choice == 4:
    total_cost = quantity * 10
    item = "Transistor"
else:
    total_cost = 0
    item = "Invalid choice"

print("\nItem Selected :", item)
print("Quantity      :", quantity)
print("Total Cost ₹  :", total_cost)
```

```python
doc_name = input("Enter document name: ")
doc_type = input("Enter document type: ").lower()

if doc_type == "pdf":
    category = "Portable Document"
elif doc_type == "doc" or doc_type == "docx":
    category = "Word Document"
elif doc_type == "xls" or doc_type == "xlsx":
    category = "Spreadsheet"
elif doc_type == "ppt" or doc_type == "pptx":
    category = "Presentation"
elif doc_type == "txt":
    category = "Text File"
elif doc_type == "jpg" or doc_type == "png":
    category = "Image File"
else:
    category = "Unknown Document Type"

print("\nDocument Name :", doc_name)
print("Document Type :", doc_type)
print("Category      :", category)
```

```python
while True:
    print("\n----- Calculator Menu -----")
    print("1. Addition")
    print("2. Subtraction")
    print("3. Multiplication")
    print("4. Division")
    print("5. Exit")

    choice = int(input("Enter your choice (1-5): "))

    if choice == 5:
        print("Calculator closed")
        break

    if choice >= 1 and choice <= 4:
        num1 = float(input("Enter first number: "))
        num2 = float(input("Enter second number: "))

        if choice == 1:
            result = num1 + num2
            print("Result:", result)

        elif choice == 2:
            result = num1 - num2
            print("Result:", result)

        elif choice == 3:
            result = num1 * num2
            print("Result:", result)

        elif choice == 4:
            if num2 != 0:
                result = num1 / num2
                print("Result:", result)
            else:
                print("Division by zero not allowed")

    else:
        print("Invalid choice")

```
